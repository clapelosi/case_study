```python
def split_words(s, separators):
    for sep in separators:
        s = s.replace(sep, ' ' + sep + ' ')
    return s.split()

def transform_string(s, dictionary):
    words = split_words(s, dictionary.keys())
    return ' '.join(dictionary.get(w, w) for w in words)

dictionary = {
    'bell': 'bello',
    'giorn': 'giorno',
    'che': 'che',
}

a = 'chebellgiorno'
a = transform_string(a, dictionary)
print(a)
```

```python
import pyitlib

def split_words(s, italian_words):
    words = []
    for word in s:
        if word in italian_words:
            words.append(word)
    return words

italian_words = pyitlib.lexicon.ItalianWords()

a = 'chebellgiorno'

a = split_words(a, italian_words)
print(a)






```
